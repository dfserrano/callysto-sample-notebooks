###### 9. In what class did the majority of the passengers on your lifeboat travel? 

```python
import csv
def majority(x, y, z):
    Max = x
    if y > Max:
        Max = y    
    if z > Max:
        Max = z
        if y > z:
            Max = y
    return Max
with open('titaniclist4.csv') as csvfile:
    first,second,third = 0,0,0
    for row in csv.DictReader(csvfile):
        if row["boat"] == "14" and row["pclass"] == "1":
            first += 1
        elif  row["boat"] == "14" and row["pclass"] == "2":
            second += 1
        elif  row["boat"] == "14" and row["pclass"] == "3":
            third += 1
        else:
            pass
        
if majority(first,second, third ) == first :
    print("The majority of passengers on my boat travelled First class")
elif majority(first,second, third ) == second :
    print("The majority of passengers on my boat travelled Second Class")
elif majority(first,second, third ) == third :
    print("The majority of passengers on my boat travelled Third Class")
else:
    print("There was no majority on my boat. There was a tie!")
```

###### Identify male and female passengers based on the salutation prefix

```python
import csv
with open('passengersinboats.csv') as csvfile:

    for row in csv.DictReader(csvfile):
        name = row["Name"]
        lname = str(name)
        if lname.startswith('Mr.'):
            print("{} was male".format(lname))
        elif lname.startswith('Sir'):
            print("{} was male".format(lname))
        elif lname.startswith('Master'):
            print("{} was male".format(lname))
        elif lname.startswith('Lt.'):
            print("{} was male".format(lname))
        elif lname.startswith('Mrs.'):
            print("{} was female".format(lname))
        elif lname.startswith('Lady'):
            print("{} was female".format(lname))
        elif lname.startswith('Miss'):
            print("{} was female".format(lname))
        elif lname.startswith('Cmdr.'):
            print("{} was male".format(lname))
        else:
             print("{}'s sex cannot be identified based on salutation prefix".format(lname))
            







        
```
